## Приоритизация гипотез

```python
#Применим фреймворк ICE для приоритизации гипотез
hypothesis['ICE']=round(hypothesis.Impact*hypothesis.Confidence/hypothesis.Efforts)
hypothesis[['Hypothesis','ICE']].sort_values(by='ICE',ascending=False)
```

Наиболее перспективные гипотезы по ICE: Запустить акцию, дающую скидку на товар в день рождения,добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей и добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок	

```python
#Применим фреймворк RICE для приоритизации гипотез
hypothesis['RICE']=round(hypothesis.Impact*hypothesis.Confidence*hypothesis.Reach/hypothesis.Efforts)
hypothesis[['Hypothesis','RICE']].sort_values(by='RICE',ascending=False)
```

Наиболее перспективные гипотезы по RICE:Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок,добавить блоки рекомендаций товаров на сайт интернет магазина, чтобы повысить конверсию и средний чек заказа и добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей	

Приоритизация гипотез при применении RICE отличается от приоритизация гипотез при применении ICE, потому что включает в себя параметр Reach, то есть скольких пользователей затронет гипотеза. Таким образом , мы определили лидирующие гипотезы в зависимости от применяемого фреймворка:

### График кумулятивной выручки по группам

```python
#приведем колонку date в orders к типу datetime
orders['date'] = orders['date'].map(
    lambda x: dt.datetime.strptime(x, '%Y-%m-%d')
)
```

```python
#приведем колонку date в visitors к типу datetime
visitors['date'] = visitors['date'].map(
    lambda x: dt.datetime.strptime(x, '%Y-%m-%d')
)
```

```python
#проверим, сколько групп в А/В тесте 
orders.group.nunique()
```

```python
#выведем дату начала теста
orders.date.min()
```

Получается, тест шел ровно 1 месяц - август 2019

```python
#выведем количество уникальных пользователей в каждой группе
orders.groupby('group')\
      .agg({'visitorId':'nunique'})
```

```python
#найдем количество уникальных групп на каждого пользователя 
filtered= (
    orders.groupby('visitorId',as_index=False)
                 .agg({'group':'nunique'})                 
)
filtered.columns=['userId', 'group_qnt']
```

```python
#check
filtered.head(10)
```

```python
#найдем пользователей, которые участвуют в нескольких группах теста одновременно
doubled_id=filtered.query('group_qnt>1')
```

```python
#check
doubled_id.head(10)
```

```python
#исключим из датасета orders пользователей, которые участвуют в нескольких группах теста одновременно
cond = orders['visitorId'].isin(doubled_id['userId'])
orders.drop(orders[cond].index, inplace = True)
```

```python
#check
orders.groupby('visitorId',as_index=False)\
                 .agg({'group':'nunique'})\
                 .query('group>1')
```

Таким образом, мы смогли выявить пользователей, которые совершали целевые действия в двух группах одновременно, было принято решение удалить таких пользователей из данных, как как их наличие не позволяет нам провести корректный AB тест и анализ его результатов , это оказывает влияние на основные метрики, которые мы рассматриваем и на основе которые формулируем результаты нашего AB теста

```python
# создаем массив уникальных пар значений дат и групп теста
datesGroups = orders[['date','group']].drop_duplicates()
```

```python
#создадим переменную ordersAggregated,содержащую:
#группу A/B-теста;
#число уникальных заказов в группе теста по указанную дату включительно;
#число уникальных пользователей, совершивших хотя бы 1 заказ в группе теста по указанную дату включительно;
#суммарную выручку заказов в группе теста по указанную дату включительно
#и отсортируем ordersAggregated по 'date' и 'group'
ordersAggregated = datesGroups.apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])]\
.agg({'date' : 'max', 'group' : 'max', 'transactionId' : 'nunique', 'visitorId' : 'nunique', 'revenue' : 'sum'}), axis=1).sort_values(by=['date','group'])
```

```python
#check
ordersAggregated.head(10) 
```

```python
#создадим переменную visitorsAggregated, в которую занесем информацию о дате, группе A/B-теста и количестве уникальных посетителей в группе теста по указанную дату включительно
visitorsAggregated = datesGroups.apply(lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])].agg({'date' : 'max', 'group' : 'max', 'visitors' : 'sum'}), axis=1).sort_values(by=['date','group'])
```

```python
#check
visitorsAggregated.head(10)
```

```python
#создадим переменную cumulativeData, объединив visitorsAggregated с ordersAggregated по 'date' и 'group'
cumulativeData = ordersAggregated.merge(visitorsAggregated, left_on=['date', 'group'], right_on=['date', 'group'])
cumulativeData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']
```

```python
#check
cumulativeData.head(10)
```

```python
#создадим переменную cumulativeRevenueA, в которую занесем данные по дате, выручке и заказам для группы А
cumulativeRevenueA = cumulativeData[cumulativeData['group']=='A'][['date','revenue', 'orders']]
```

```python
#создадим переменную cumulativeRevenueB, в которую занесем данные по дате, выручке и заказам для группы B
cumulativeRevenueB = cumulativeData[cumulativeData['group']=='B'][['date','revenue', 'orders']]
```

```python
#построим график кумулятивной выручки по группам 
title = plt.title('График кумулятивной выручки по группам')
xlabel = plt.xlabel('Дата')
ylabel = plt.ylabel('Выручка')
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue'], label='A')
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue'], label='B')
plt.legend();
```

Можем отметить, что в течение всего периода проведения теста выручки обеих групп растут, выручка группы B больше, чем группы А, также можно заметить, что график выручки группы B в нескольких местах не плавный, скорее всего на это повлияли выбросы и аномальные значения

### График кумулятивного среднего чека по группам

```python
#аналогично построим график кумулятивного среднего чека по группам
title = plt.title('График кумулятивного среднего чека по группам')
xlabel = plt.xlabel('Дата')
ylabel = plt.ylabel('Средний чек')
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue']/cumulativeRevenueA['orders'], label='A')
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue']/cumulativeRevenueB['orders'], label='B')
plt.legend();
```

Средний чек группы B значительно превышает средний чек группы А, на этом графике присутствует больше точек резкого роста и падения, что также говорит о наличии аномалий и выбросов

### Cтатистическая значимость различий в среднем количестве заказов на посетителя между группами по «сырым» данным

Перед тем, как перейти к подсчету переменных, следует выдвинуть гипотезы:

- H0: Различий в среднем количестве заказов на посетителя между группами нет
- H1: Различия в среднем количестве заказов на посетителя между группами есть

В работе мы будем использовать непараметрический тест Уилкоксона-Манна-Уитни, это структурный подход, он используется для данных с большими(по сравнению с нормальным распределением) выбросами, какими и являются наши данные

```python
#создадим переменную visitorsADaily, содержащую данные по датам и количеству визитов пользователей по дням группы A
visitorsADaily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA']
```

```python
#создадим переменную visitorsBDaily, содержащую данные по датам и количеству визитов пользователей по дням группы B
visitorsBDaily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB']
```

```python
#создадим перменную ordersByUsersA,посчитаем количество заказов на каждого пользователя группы А
ordersByUsersA = (
    orders[orders['group'] == 'A']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique})
)
ordersByUsersA.columns = ['userId', 'orders']
```

```python
#создадим перменную ordersByUsersB,посчитаем количество заказов на каждого пользователя группы B
ordersByUsersB = (
    orders[orders['group'] == 'B']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique})
)
ordersByUsersB.columns = ['userId', 'orders']
```

```python
#создадим переменную sampleA, в которой будет информация о количестве заказов на пользователя, также для тех пользователей, которые ничего не покупали, будет стоять 0 для группы А
sampleA = pd.concat(
    [
        ordersByUsersA['orders'],
        pd.Series(
            0,
            index=np.arange(
                visitorsADaily['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
```

```python
#создадим переменную sampleB, в которой будет информация о количестве заказов на пользователя, также для тех пользователей, которые ничего не покупали, будет стоять 0 для группы B
sampleB = pd.concat(
    [
        ordersByUsersB['orders'],
        pd.Series(
            0,
            index=np.arange(
                visitorsBDaily['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
```

```python
#найдем p-value 
'p-value='"{0:.3f}".format(stats.mannwhitneyu(sampleA, sampleB)[1])
```

```python
#Выведем относительный прирост среднего числа заказов группы B
'относительная разница='"{0:.3f}".format(sampleB.mean() / sampleA.mean() - 1)
```

p-value=0.011, это меньше ,чем 0.05, значит , отвергаем H0, различия в среднем количестве заказов на посетителя между группами есть.
Выигрыш группы B составляет 16%

### Статистическая значимость различий в среднем чеке заказа между группами по «сырым» данным

Сформируем гипотезы:
H0: Различий в среднем чеке заказа между группами нет
H1: Различия в среднем чеке заказа между группами есть

```python
#найдем p-value
'p-value=''{0:.3f}'.format(stats.mannwhitneyu(orders[orders['group']=='A']['revenue'], orders[orders['group']=='B']['revenue'])[1])
```

```python
#Выведем относительный прирост среднего чека заказа группы B
'относительная разница=''{0:.3f}'.format(orders[orders['group']=='B']['revenue'].mean()/orders[orders['group']=='A']['revenue'].mean()-1)
```

p-value=0.829, это больше, чем 0.05, поэтому нет оснований отвергать H0,различий в среднем чеке заказа между группами нет

### Статистическая значимость различий в среднем количестве заказов на посетителя между группами по «очищенным» данным

```python
#создадим переменную usersWithManyOrders, в которой отфильтруем изначальные данные в соответствии с решением об аномальных пользователях
usersWithManyOrders = pd.concat(
    [
        ordersByUsersA[ordersByUsersA['orders'] > np.percentile(ordersByUsers['orders'], 95)]['userId'],
        ordersByUsersB[ordersByUsersB['orders'] > np.percentile(ordersByUsers['orders'], 95)]['userId'],
    ],
    axis=0,
)
usersWithExpensiveOrders = orders[orders['revenue'] > np.percentile(orders['revenue'],95)]['visitorId']


abnormalUsers = (
    pd.concat([usersWithManyOrders, usersWithExpensiveOrders], axis=0)
    .drop_duplicates()
    .sort_values()
)
```

```python
#проверим количество аномальных пользователей
abnormalUsers_qnt=abnormalUsers.shape[0]
```

```python
abnormalUsers_qnt
```

```python
#проверим процентр относительных потерь 
abnormalUsers_qnt/orders['visitorId'].nunique()
```

Получается, примерно 9% покупателей совершили аномальные покупки 

```python
#создадим переменную sampleAFiltered по отфильтрованным данным, содержащую информацию о количестве заказов на пользователя, также для тех пользователей, которые ничего не покупали, будет стоять 0 для группы А
sampleAFiltered = pd.concat(
    [
        ordersByUsersA[
            np.logical_not(ordersByUsersA['userId'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                visitorsADaily['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
```

```python
#создадим переменную sampleBFiltered по отфильтрованным данным, содержащую информацию о количестве заказов на пользователя, также для тех пользователей, которые ничего не покупали, будет стоять 0 для группы B
sampleBFiltered = pd.concat(
    [
        ordersByUsersB[
            np.logical_not(ordersByUsersB['userId'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                visitorsBDaily['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
) 
```

```python
#найдем p-value
'p-value=''{0:.3f}'.format(stats.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1])
```

```python
#Выведем относительный прирост среднего числа заказов группы B
'относительная разница=''{0:.3f}'.format(sampleBFiltered.mean()/sampleAFiltered.mean()-1)
```

p-value=0.016, прирост 17.4% - результаты по количеству заказов на посетителя остались примерно на том же уровне, различия в среднем количестве заказов на посетителя между группами есть

### Статистическая значимость различий в среднем чеке заказа между группами по «очищенным» данным

```python
'p-value=''{0:.3f}'.format(
        stats.mannwhitneyu(
            orders[
                np.logical_and(
                    orders['group'] == 'A',
                    np.logical_not(orders['visitorId'].isin(abnormalUsers)),
                )
            ]['revenue'],
            orders[
                np.logical_and(
                    orders['group'] == 'B',
                    np.logical_not(orders['visitorId'].isin(abnormalUsers)),
                )
            ]['revenue'],
        )[1]
    )
```

```python
'относительная разница='"{0:.3f}".format(
        orders[
            np.logical_and(
                orders['group'] == 'B',
                np.logical_not(orders['visitorId'].isin(abnormalUsers)),
            )
        ]['revenue'].mean()
        / orders[
            np.logical_and(
                orders['group'] == 'A',
                np.logical_not(orders['visitorId'].isin(abnormalUsers)),
            )
        ]['revenue'].mean()
        - 1
    )
```

p-value увеличился, в то же самое время разница между сегментами сократилась,различий в среднем чеке заказа между группами нет

Исходя из обнаруженных фактов, тест следует остановить и признать победу группы B
