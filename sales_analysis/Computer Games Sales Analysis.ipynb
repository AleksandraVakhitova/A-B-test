# Study on the video games sales

## The purpose of the study:
Identify the patterns that determine the success of the video game, which will allow to choose a potentially popular product and plan future advertising campaigns

## Objectives of the study:
-Data preparationd: data transformation, gaps analysis
-Exploratory data analysis:
to analyze sales by platform during the given period
to determine the relevant period for further analysis, depending on the results of previous steps
to identify several potentially profitable platforms during the relevant period
to analyze global game sales by platform during the relevant period
to analyze the correlation between platform sales and user and critic reviews over the relevant period
to analyze the overall distribution of games by genre, identifying genres with high and low sales during the relevant period
-Create user profiles for each region:
to determine the most popular platforms (top 5) for each region during the relevant period
to identify the most popular genres (top 5) for each region during the relevant period
to determine whether the ESRB rating affects sales in a particular region
-Hypothesis testing:
1:the average user ratings of the Xbox One and PC platforms are the same
2:average user ratings for Action and Sports genres are different

## Data used in the study
Historical game sales data, user and expert ratings, genres and platforms

```python
#importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats as st
```

```python
#reading the content of the file
data = pd.read_csv('games.csv')
```

```python
#displaying the first 10 rows
data.head(10)
```

We received the data, we can seee that the data contains information about the game name, platform, release year, genre, sales in North America, Europe, Japan and other regions, as well as critic and user rating and ESRB rating

## Data preprocessing 

### Changing column names

```python
#changing column names to lower case
data.columns = data.columns.str.lower()
```

```python
#check
data.head(10)
```

### Working with missing data

```python
#identifying gaps 
data.isnull().sum()
```

```python
#getting rid of gaps in the 'name' column as the number of gaps is small (2 rows)
data = data.dropna(subset=['name'])
```

```python
#getting rid of gaps in the 'year_of_release' column as the number of gaps is small (269 rows)
data = data.dropna(subset=['year_of_release'])
```

I decided not to replace the gaps in the columns "critic_score", "user_score", "rating" , as replacing them with the median, mean or 0 will greatly affect further research, the data will be distorted. There are 2 main reasons that could lead to gaps: 1) Error unloading (collecting) data from the system 2) Human factor

```python
#tbd means to be determined, most likely it means that the user has not yet decided on the rating, for further research we will replace tbd with NaN
data['user_score'] = data['user_score'].replace('tbd','NaN')
```

```python
#check
data.isnull().sum()
```

### Changing the data type

```python
#examining the type of data provided 
data.info()
```

```python
#changing the data type in the 'year_of_release' column to int
data['year_of_release'] = data['year_of_release'].astype('int')
```

```python
#changing the data type in the 'user_score' column to float
data['user_score'] = pd.to_numeric(data['user_score'], errors='coerce')
```

```python
#changing the data type in the 'critic_score' column to int and filling gaps in the column 'critic_score' with -999
data['critic_score'] = data['critic_score'].fillna (-999)
data['critic_score'] = data['critic_score'].astype('int')
```

```python
#finding duplicates
data.duplicated().sum()
```

```python
#check
data.info()
```

## Hypothesis testing

### Hypothesis about the equality of the average user ratings of the Xbox One and PC platforms

```python
#H0: The average user ratings of the Xbox One and PC platforms are the same
#H1: Average user ratings for Xbox One and PC platforms are different
#alpha = 0.05


Xbox_One = new_data[new_data['platform'] =='XOne']['user_score']
PC = new_data[new_data['platform'] =='PC']['user_score']

results = st.ttest_ind(Xbox_One, PC, equal_var = False,nan_policy='omit')
alpha = 0.05

print(results.pvalue)
if results.pvalue < alpha:
    print('Rejecting the Null Hypothesis')
else:
    print('Failing to reject a null hypothesis')
```

The task is to test the hypothesis that the average user ratings of the Xbox One and PC platforms are the same, this will be the null hypothesis H0, which means that the alternative hypothesis H1 will state that the average user ratings of the Xbox One and PC platforms are different. We used the method to test the hypothesis about the equality of the mean of two general populations for the samples taken from them with the False parameter, that is, we believe that the variances of the samples are different, we also used the standard alpha of 5%.
As a result, we got a test result of 0.54, so we cannot reject the null hypothesis: the average user ratings of the Xbox One and PC platforms are indeed the same.

### Hypothesis about the difference between the average user ratings of Action and Sports genres

```python
#H0:Average user ratings for Action and Sports are the same
#H1:Average user ratings for Action and Sports are different
#alpha = 0.05


action = new_data[new_data['genre'] =='Action']['user_score']
sports = new_data[new_data['genre'] =='Sports']['user_score']

results = st.ttest_ind(action, sports, equal_var = False,nan_policy='omit')
alpha = 0.05

print(results.pvalue)
if results.pvalue < alpha:
    print('Rejecting the Null Hypothesis')
else:
    print('Failing to reject a null hypothesis')
```

Another task is to test the hypothesis that the average user ratings of the Action and Sports genres are different, this will be the alternative hypothesis H1, which means the null hypothesis H0 will state that the average user ratings of the Action and Sports genres are the same. Here we also used the method to test the hypothesis about the equality of the mean of two general populations for the samples taken from them with the False parameter, that is, we believe that the variances of the samples are different, we also used the standard alpha of 5%.
As a result, we got a test result of 4.2, so we can reject the null hypothesis: the average user ratings of the Action and Sports genres are indeed different.
